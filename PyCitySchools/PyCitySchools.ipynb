# PyCity Schools Analysis

The analysis of testing scores in the areas of Maths and Reading is presented below:

**Overall Assessment:**
Based on the assessment of 39,170 students from 15 different schools, the initial findings reveal that the average score for math was 70.34, while for reading it was 69.98. Additionally, the overall passing rate for math was 86.08%, and for reading it was 84.43%. Finally, 72.81% of the students were able to pass both tests.

**School Performance:**
Griffin High School has the highest overall passing rate at 81.34%, exceeding the average by almost 9%. In contrast, Hernandez High School exhibited the lowest performance, with an overall passing rate of only 66.36%.

**Test Scores by Year:**
The results for the average score on both Maths and Reading by year do not show any significant differences between them. The test scores by year are all very close to their respective global average without any sign of outliers.

**Budget per Capita:**
Surprisingly, the best overall passing results when looking at the budget per capita do not belong to the students in which more budget is spent ($645-680) but rather to the $585-630 spending range followed by the lowest spending range, with results of 79.88% and 76.72% respectively. Furthermore, the schools with the fewest number of students (<1,000) presented a considerably much higher overall passing rate than the large schools.

**Type of School:**
Lastly, a final view of the overall test scores by type of school reveals that Independent Schools have much higher results in both Maths and Reading Test Scores than Government Schools.

**Conclusions:**
From the above analysis, two important conclusions can be drawn:
* In average, Maths Test Scores are slightly higher than Reading Test Scores.
* Independent Schools have much better test scores overall and by subject than Government Schools, with 3 Independent Schools in the Top 5 performing Schools, and 4 Government Schools in the Bottom 5 performing Schools.

```python
# Dependencies and Setup
from pathlib import Path
import pandas as pd

# Source Files Paths
schools_file_path = Path("Resources/schools_complete.csv")
students_file_path = Path("Resources/students_complete.csv")

# Dataframe Creation
schools_df = pd.read_csv(schools_file_path)
students_df = pd.read_csv(students_file_path)

# Combine Dataframes
main_df = pd.merge(students_df, schools_df, how="left", on=["school_name", "school_name"])
main_df
```

## Local Government Area  (LGA) Summary

```python
# Totals (School and Students)
school_count = main_df["school_name"].nunique()
student_count = main_df["student_name"].count()

# Total Budget
total_budget = schools_df["budget"].sum()

# Average Scores
avg_maths_score = main_df["maths_score"].mean()
avg_reading_score = main_df["reading_score"].mean()

# Percentage Passing Rates
passing_maths_count = main_df[(main_df["maths_score"]>=50)].count()["student_name"]
passing_reading_count = main_df[(main_df["reading_score"]>=50)].count()["student_name"]
overall_passing_count = main_df[(main_df["maths_score"]>=50) & (main_df["reading_score"]>=50)].count()["student_name"]
passing_maths_percentage = (passing_maths_count / float(student_count)) * 100
passing_reading_percentage = (passing_reading_count / float(student_count)) * 100
overall_passing_percentage = (overall_passing_count / float(student_count)) * 100
```

```python
# LGA Summary DataFrame
area_summary = pd.DataFrame({
    "Total Schools": [school_count],
    "Total Students": [student_count],
    "Total Budget": [total_budget],
    "Average Maths Score": [avg_maths_score],
    "Average Reading Score": [avg_reading_score],
    "% Passing Maths": [passing_maths_percentage],
    "% Passing Reading": [passing_reading_percentage],
    "% Overall Passing": [overall_passing_percentage]})

# Formatting
area_summary["Total Students"] = area_summary["Total Students"].map("{:,}".format)
area_summary["Total Budget"] = area_summary["Total Budget"].map("${:,.2f}".format)

# Display the Summary of DataFrame
area_summary
```

## School Summary

```python

# School Type per School Name
school_types = schools_df.set_index(["school_name"])["type"]

# Total Students per School Name
per_school_counts = schools_df.set_index(["school_name"])["size"]

# Total Budged per School Name and Budget per Capita
per_school_budget = schools_df.set_index(["school_name"])["budget"]
per_school_capita = per_school_budget / per_school_counts

# Average Test Scores (Maths and Reading) per School Name
per_school_maths = main_df.groupby("school_name")["maths_score"].mean()
per_school_reading = main_df.groupby("school_name")["reading_score"].mean()
```

```python
# Filtered DataFrames for Passing Grades (Maths and Reading)
school_passing_maths = main_df[main_df["maths_score"]>=50]
school_passing_reading = main_df[main_df["reading_score"]>=50]

# Filtered DataFrames for Passing Grades (Overall)
passing_maths_and_reading = main_df[(main_df["maths_score"]>=50) & (main_df["reading_score"]>=50)]
```

```python
# Pass Rates Percentages (Maths, Reading and Overall) per School Name
per_school_passing_maths = (school_passing_maths.groupby("school_name")["student_name"].count() / per_school_counts) * 100
per_school_passing_reading = (school_passing_reading.groupby("school_name")["student_name"].count() / per_school_counts) * 100
overall_passing_rate = (passing_maths_and_reading.groupby("school_name")["student_name"].count() / per_school_counts) * 100
```

```python
# Summary DataFrame per School Name
per_school_summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": per_school_counts,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_school_capita,
    "Average Maths Score": per_school_maths,
    "Average Reading Score": per_school_reading,
    "% Passing Maths": per_school_passing_maths,
    "% Passing Reading": per_school_passing_reading,
    "% Overall Passing": overall_passing_rate})

# DataFrame Formatting
per_school_summary_formatted = per_school_summary.copy()
per_school_summary_formatted["Total School Budget"] = per_school_summary_formatted["Total School Budget"].map("${:,.2f}".format)
per_school_summary_formatted["Per Student Budget"] = per_school_summary_formatted["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary_formatted
```

## Top Performing Schools (By % Overall Passing)

```python
# Top 5 Schools
top_schools = per_school_summary.sort_values("% Overall Passing", ascending=False)
top_schools.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

```python
# Bottom 5 Schools
bottom_schools = per_school_summary.sort_values("% Overall Passing", ascending=True)
bottom_schools.head(5)
```

## Maths Scores by Year

```python
# DataFrame Filtering by Year
year_nine = main_df[(main_df["year"] == 9)]
year_ten = main_df[(main_df["year"] == 10)]
year_eleven = main_df[(main_df["year"] == 11)]
year_twelve = main_df[(main_df["year"] == 12)]

# Data Series for the Average Maths Score by School Name per Year
year_nine_math_scores = year_nine.groupby("school_name")["maths_score"].mean().round(2)
year_ten_math_scores = year_ten.groupby("school_name")["maths_score"].mean().round(2)
year_eleven_math_scores = year_eleven.groupby("school_name")["maths_score"].mean().round(2)
year_twelve_math_scores = year_twelve.groupby("school_name")["maths_score"].mean().round(2)

# DataFrame for the Average Maths Score by School Name per Year
maths_scores_by_year = pd.DataFrame({
    "Year 9": year_nine_math_scores,
    "Year 10": year_ten_math_scores,
    "Year 11": year_eleven_math_scores,
    "Year 12": year_twelve_math_scores})

# DataFrame Index Name Removal
maths_scores_by_year.index.name = None

# Display the DataFrame
maths_scores_by_year
```

## Reading Scores by Year

```python
# Data Series for the Average Reading Score by School Name per Year
year_nine_reading_scores = year_nine.groupby("school_name")["reading_score"].mean().round(2)
year_ten_reading_scores = year_ten.groupby("school_name")["reading_score"].mean().round(2)
year_eleven_reading_scores = year_eleven.groupby("school_name")["reading_score"].mean().round(2)
year_twelve_reading_scores = year_twelve.groupby("school_name")["reading_score"].mean().round(2)

# DataFrame for the Average Reading Score by School Name per Year
reading_scores_by_year = pd.DataFrame({
    "Year 9": year_nine_reading_scores,
    "Year 10": year_ten_reading_scores,
    "Year 11": year_eleven_reading_scores,
    "Year 12": year_twelve_reading_scores})

# DataFrame Index Name Removal
reading_scores_by_year.index.name = None

# Display the DataFrame
reading_scores_by_year
```

## Scores by School Spending

```python
# Copy of School Summary DataFrame
school_spending_df = per_school_summary
```

```python
# School Spending Bins Definition
spending_bins = [0, 585, 630, 645, 680]
group_names_spending = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create New Column on Dataframe with Spending Bin
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], spending_bins, labels=group_names_spending, include_lowest=True)
school_spending_df
```

```python
# Series Definition for the Average Scores and Passing Percentages by Spending Range
spending_maths_scores = school_spending_df.groupby("Spending Ranges (Per Student)", observed=False)["Average Maths Score"].mean().round(2)
spending_reading_scores = school_spending_df.groupby("Spending Ranges (Per Student)", observed=False)["Average Reading Score"].mean().round(2)
spending_passing_maths = school_spending_df.groupby("Spending Ranges (Per Student)", observed=False)["% Passing Maths"].mean().round(2)
spending_passing_reading = school_spending_df.groupby("Spending Ranges (Per Student)", observed=False)["% Passing Reading"].mean().round(2)
overall_passing_spending = school_spending_df.groupby("Spending Ranges (Per Student)", observed=False)["% Overall Passing"].mean().round(2)
```

```python
# DataFrame Definition for the Average Scores and Passing Percentages by Spending Range
spending_summary = pd.DataFrame({
    "Average Math Score": spending_maths_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Maths": spending_passing_maths,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending})

# Display results
spending_summary
```

```python
# DataFrame Definition for the Average Scores and Passing Percentages by School Size
size_summary = pd.DataFrame({
    "Average Maths Score": size_maths_scores,
    "Average Reading Score": size_reading_scores,
    "% Passing Maths": size_passing_maths,
    "% Passing Reading": size_passing_reading,
    "% Overall Passing": size_overall_passing})

# Display results
size_summary
```
